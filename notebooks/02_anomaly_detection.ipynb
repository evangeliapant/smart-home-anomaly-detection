```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load processed features
df = pd.read_csv("../data/processed/hh101_features_10T.csv", index_col=0, parse_dates=True)

# Scale and cluster
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

kmeans = KMeans(n_clusters=6, random_state=42)
df["cluster"] = kmeans.fit_predict(X_scaled)

# Visualize clusters
plt.figure(figsize=(12,5))
plt.plot(df.index, df.sum(axis=1), color="gray", alpha=0.5)
plt.scatter(df.index, df.sum(axis=1), c=df["cluster"], cmap="tab10", s=10)
plt.title("Detected Routines (K-Means Clusters)")
plt.show()
```
